```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

%matplotlib inline
```

```python
churndata = pd.read_csv("data/churn-data.csv")
churndata.isna().sum()
```

```python
churndata.credit_score.fillna(churndata.credit_score.mean(), inplace = True)
churndata.rewards_earned.fillna(churndata.rewards_earned.mean(), inplace = True)
churndata.age.fillna(churndata.age.mean(), inplace = True)
churndata.rewards_earned
```

```python
churndata['zodiac_sign'] = churndata['zodiac_sign'].replace('na',np.nan)
churndata['housing'] = churndata['housing'].replace('na',np.nan)
churndata['payment_type'] = churndata['payment_type'].replace('na',np.nan)

churndata['zodiac_sign'].fillna(churndata['zodiac_sign'].mode().values[0], inplace = True)
churndata['housing'].fillna(churndata['housing'].mode().values[0], inplace = True)
churndata['payment_type'].fillna(churndata['payment_type'].mode().values[0], inplace = True)
```

```python
churndata.drop(['user'], axis=1, inplace=True)
```

```python
for i in churndata.columns:
    if i != "churn" and churndata[i].corr(churndata.churn)<0:
        churndata.drop([i], axis=1, inplace=True)
        print(i + " was removed.")
```

```python
churndata
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(churndata["churn"])
print(y)
standardisation = MinMaxScaler()  
# Scaled feature 
churndata = standardisation.fit_transform(churndata) 
  
print ("\nAfter Standardisation : \n", churndata) 

X = churndata[:,1:]
X, y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=True)
print("X_train", len(X_train))
print("X_test", len(X_test))
print("y_train", len(y_train))
print("y_test", len(y_test))
```

```python
#get number of columns in training data
n_cols = X_train.shape[1]

model = Sequential()
model.add(Dense(200, activation='relu', input_shape=(n_cols,)))
model.add(Dense(200, kernel_initializer='normal', activation='relu'))
model.add(Dense(2, kernel_initializer='normal', activation='softmax'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])  # for binary classification
```

```python
history = model.fit(X_train, y_train, epochs=200)
```

```python
classes = model.predict(X_test)
classes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# Plot history: MAE
plt.plot(history.history['accuracy'], label='accuracy (training data)')
plt.title('Churn Data')
plt.ylabel('Accuracy')
plt.xlabel('No. epoch')
plt.legend(loc="upper left")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf_forest = RandomForestClassifier(n_estimators=100, max_depth=10)
clf_forest.fit(X_train, y_train)

pred = clf_forest.predict(X_train)
print("Training accuracy", accuracy_score(y_train, pred))

pred_test = clf_forest.predict(X_test)
print("Test accuracy",accuracy_score(y_test, pred_test))
```
